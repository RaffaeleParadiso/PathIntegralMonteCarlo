```python
def distanza(x, y):
    diff = y - x
    if (diff > 0.5):
        return diff - 1
    if (diff < -0.5):
        return 1 + diff
    else:
        return diff
```

```python
def avvolgimento(Nt, y, g):
    sum = 0
    for i in range(0, Nt):
        sum +=(distanza(g[i], y[i]))
    return round(sum)
```
